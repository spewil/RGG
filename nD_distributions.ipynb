```python
%matplotlib inline
import numpy as np
import network_generation.generation as netgen
import matplotlib.pyplot as plt
import experiment as ex
import time 

# PARAMS 
kappa, n, d = 10, 1000, 2
```

```python
## SOLID  
RGGs = netgen.RGGEnsemble(kappa, n, d,boundary='s',num_radii=10)
RGGs.generate_samples(n=100)
RGGs.samples[0].plot_network(unmatched = RGGs.samples[0].find_unmatched())
print RGGs.get_num_samples()
print np.mean([sample.mean_degree() for sample in RGGs.samples])
print RGGs.radius
print RGGs.radius_std
```

```python
nD_list = [sample.find_num_unmatched() for sample in RGGs.samples]
```

```python
# nD distribution
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
n, bins, patches = ax1.hist(nD_list,19, normed=1, facecolor='red', alpha=0.75)
ax1.set_title('$n_D$ distribution solid BC')
ax1.set_xlabel('$n_D$')
ax1.set_ylabel('frequency')
print 'mean nD: ' + str(np.mean(nD_list))
fig1.savefig('./plots/histograms/nD_100samples_'+ RGGs.get_param_string())
```
