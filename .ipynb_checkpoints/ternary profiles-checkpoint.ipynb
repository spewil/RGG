```python
# validate (s,e,i) with a small graph 
boundary = 's'
kappa = 3
n = 5
d = 2

RGG = ex.RGGEnemble(kappa, n, d, shortcut_prob=0, boundary=boundary, num_radii=10)
RGG.generate_samples(n=1)
s = RGG.samples[0]:
    sz = zen.DiGraph.from_adj_matrix(np.array(s.adjacency_dense))
    profile_list.append(zen.control.profile(sz))
```

```python
def add_corner_labels(fig,fntsz=20):
    fig.text(.48,.92,'$n_D^i$', fontsize=fntsz )
    fig.text(.17,.08,'$n_D^s$', fontsize=fntsz )
    fig.text(.8,.08,'$n_D^e$', fontsize=fntsz )
    
def save_ternary_plots(colors,markers):
    i = 0
    total_profile_list = []
    d_profile_list = []
    for d in d_range:
        profile_list = []
        RGG = ex.RGGExperiment(kappa_range, n, d, shortcut_prob=0, boundary=boundary, num_radii=10)
        for ensemble in RGG.ensembles:
            for s in ensemble.samples:
                sz = zen.DiGraph.from_adj_matrix(np.array(s.adjacency_dense))
                profile_list.append(zen.control.profile(sz))

        # add all d things together 
        d_profile_list.append(profile_list)
        total_profile_list.extend(profile_list)

    for i in range(len(d_range)):
        fig = plt.figure(i+1)
        h = zen.control.profile_plot(d_profile_list[i],color=colors[i],marker=markers[i]) 
        add_corner_labels(fig)
        ax = fig.gca()
        ax.set_xlabel('$d = $'+str(d_range[i]))
        fig.savefig('./plots/ternary/solid/RGG_points_N_'+str(n)+'_d_'+str(d_range[i])+'_BC_'+str(boundary)+'_.eps',dpi=1000)

    for i in range(len(d_range)):
        fig = plt.figure(10+i+1)
        h = zen.control.profile_heatmap(d_profile_list[i],color=colors[i],marker=markers[i]) 
        add_corner_labels(fig)
        ax = fig.gca()
        ax.set_xlabel('$d = $'+str(d_range[i]))
        fig.savefig('./plots/ternary/solid/RGG_heatmap_N_'+str(n)+'_d_'+str(d_range[i])+'_BC_'+str(boundary)+'_.eps',dpi=1000)

    fig = plt.figure(len(d_range)+1)
    for i in range(len(d_range)):
        h = zen.control.profile_plot(d_profile_list[i],color=colors[i],marker=markers[i]) 
    add_corner_labels(fig)
    ax = fig.gca()
    ax.set_xlabel('all dimensions')
    fig.savefig('./plots/ternary/solid/RGG_total_points_N_'+str(n)+'_BC_'+str(boundary)+'_.eps',dpi=1000)

    fig = plt.figure(len(d_range)+2)
    h = zen.control.profile_heatmap(total_profile_list,color='red',marker='x') 
    add_corner_labels(fig)
    ax = fig.gca()
    ax.set_xlabel('all dimensions')
    fig.savefig('./plots/ternary/solid/RGG_total_heatmap_N_'+str(n)+'_BC_'+str(boundary)+'_.eps',dpi=1000)
    
def draw_bar_chart(d_profile_list,ER_profile_list):
    # sum across all node types and average
    # for each dimension 
    s_means = []
    s_stds = []
    e_means = []
    e_stds = []
    i_means = []
    i_stds = []

    for dl in d_profile_list:
        d_s = []
        d_e = []
        d_i = []
        # for each 3-tuple
        for t in dl:
            d_s.append(t[0])
            d_e.append(t[1])
            d_i.append(t[2])        

        s_means.append(np.mean(d_s)) 
        s_stds.append(stats.sem(d_s))

        e_means.append(np.mean(d_e)) 
        e_stds.append(stats.sem(d_e))

        i_means.append(np.mean(d_i)) 
        i_stds.append(stats.sem(d_i))
        
    # make sure they sum to 1
    for i in range(len(s_means)):
        total_fraction = s_means[i] + e_means[i] + i_means[i]
        s_means[i] = s_means[i]/float(total_fraction)
        s_stds[i] = s_stds[i]/float(total_fraction)
        
        e_means[i] = e_means[i]/float(total_fraction)
        e_stds[i] = e_stds[i]/float(total_fraction)

        i_means[i] = i_means[i]/float(total_fraction)
        i_stds[i] = i_stds[i]/float(total_fraction)

    # bar chart showing the breakdown 
    # s --> e --> i 

    fig = plt.figure(100)
    ax = fig.add_subplot(1,1,1)
    ind = d_range # the x locations for the groups
    width = [o*.8 for o in np.ones(len(d_range))] # the width of the bars: can also be len(x) sequence

    ####
    
    # ER data 
    # for each tuple
    ER_s = []
    ER_e = []
    ER_i = []
    for t in ER_profile_list:
        ER_s.append(t[0])
        ER_e.append(t[1])
        ER_i.append(t[2])
    subtotal = float(np.mean(ER_s)+np.mean(ER_e)+np.mean(ER_i))
    ER_s_mean = np.mean(ER_s)/subtotal
    ER_s_std = stats.sem(ER_s)/subtotal
    ER_e_mean = np.mean(ER_e)/subtotal
    ER_e_std = stats.sem(ER_e)/subtotal
    ER_i_mean = np.mean(ER_i)/subtotal
    ER_i_std = stats.sem(ER_i)/subtotal
    
    # ER lines 
    length = [min(d_range)-.45, max(d_range)+.45] 
    ERh = ax.plot(length,[ER_s_mean, ER_s_mean],'c-',label='ER source',zorder=-10)
    ax.plot(length,[ER_s_mean+ER_s_std, ER_s_mean+ER_s_std],'c--',zorder=-10)
    ax.plot(length,[ER_s_mean-ER_s_std, ER_s_mean-ER_s_std],'c--',zorder=-10)
    
    ERh = ax.plot(length,[ER_e_mean+ER_s_mean, ER_e_mean+ER_s_mean],'m-',label='ER external',zorder=-20)
    ax.plot(length,[ER_e_mean+ER_s_mean+ER_e_std, ER_e_mean+ER_s_mean+ER_e_std],'m--',zorder=-20)
    ax.plot(length,[ER_e_mean+ER_s_mean-ER_e_std, ER_e_mean+ER_s_mean-ER_e_std],'m--',zorder=-20)
    
    ERh = ax.plot(length,[ER_i_mean+ER_e_mean+ER_s_mean, ER_i_mean+ER_e_mean+ER_s_mean],'y-',label='ER internal',zorder=-30)
    ax.plot(length,[ER_i_mean+ER_e_mean+ER_s_mean+ER_i_std, ER_i_mean+ER_e_mean+ER_s_mean+ER_i_std],'y--',zorder=-30)
    ax.plot(length,[ER_i_mean+ER_e_mean+ER_s_mean-ER_i_std, ER_i_mean+ER_e_mean+ER_s_mean-ER_i_std],'y--',zorder=-30)
    
    ####
    
    p1 = ax.bar(ind, s_means, width, color='g', yerr=s_stds, label='source')
    p2 = ax.bar(ind, e_means, width, color='b', bottom=s_means, yerr=e_stds,label='external')
    p3 = ax.bar(ind, i_means, width, color='r', bottom=[e+s for e,s in zip(e_means,s_means)], yerr=i_stds,label='internal')
    
    ax.set_ylabel('Mean fraction of nD')
    ax.set_xticks(ind, [str(d) for d in d_range])
    ax.set_yticks(np.arange(0, 1.1, .1))
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles,labels)  

    return fig     
```
