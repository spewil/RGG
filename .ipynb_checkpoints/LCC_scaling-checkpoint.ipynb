## scaling for graph LCCs only

- we want to only extract scaling over $\kappa$ above the percolation threshold 
- we know what this threshold is for ER graphs 
- first show the difference in scaling for ER, RGGp, and RGGs for a single dimension 
- then show the scaling parameters over a range of dimensions 


```python
# do the full kappa_range scaling 
# do the shortened kappa_range scaling 
# plot both for ER, RGGs, RGGp

kappa_range = [1,3,5,10,15,20]
kappa_range_LCC = [10,15,20]

kappa_range_fine = np.arange(0.1,22,.01)
kappa_range_fine_LCC = np.arange(8,22,.01)

n = 100
num_radii = 10
boundaries = ['s','p']
d_range = [2] #[2,3,6,9,15,25,40]
d = 2 

# MODEL
def f(B, x):
    return np.exp(B[0]*x)
```

```python
# full range scaling 

# set up plots and axes 
ndkfig = plt.figure(1)
ndkax = ndkfig.add_subplot(1,1,1)

## ER 
ndkhandles = []
ndklabels = []

# ER data and fit
ERE = ex.ERExperiment(kappa_range, n)
# plt.errorbar(ERE.mean_degree_list,ERE.mean_nD_list,fmt='ro',markersize=3,ecolor='r',xerr=ERE.std_degree_list,yerr=ERE.std_nD_list)
mean_degree_list, std_degree_list, _ = ERE.find_degree_stats()
mean_nD_list, std_nD_list, _ = ERE.find_nD_stats()

x = mean_degree_list
y = mean_nD_list 
wx = [s+1 for s in std_degree_list]
wy = [s+1 for s in std_nD_list] 

# fitting 
# popt = optimize.curve_fit(lambda t,b:np.exp(b*t), RGG.mean_degree_list, RGG.mean_nD_list)
popt = [[-0.5]]
exponential = Model(f)
data = Data(x, y, wd=wx, we=wy)
odr = ODR(data, exponential, beta0=popt[0])
output = odr.run()
ndkhandle, = ndkax.plot(kappa_range_fine_LCC,f(output.beta,kappa_range_fine_LCC),'--')
    
ndkhandles.append(ndkhandle)
ndklabels.append('ER Simulation')

print 'ER sim done', 'gamma = ' + str(output.beta[0]) + ' +/- ' + str(output.sd_beta[0])

## RGG 
for boundary in boundaries:
    if boundary == 's':
        color = 'r'
    elif boundary == 'p':
        color = 'b'
    gammas = []
    gamma_errs = []

    RGG = ex.RGGExperiment(kappa_range, n, d, shortcut_prob=0, boundary=boundary, num_radii=10)

    mean_degree_list, std_degree_list, _ = RGG.find_degree_stats()
    mean_nD_list, std_nD_list, _ = RGG.find_nD_stats()

    x = mean_degree_list
    y = mean_nD_list

    # plt.errorbar(x, y, fmt='none', yerr=RGG.std_nD_list, xerr=RGG.std_degree_list)
    # fitting 
    # popt = optimize.curve_fit(lambda t,b:np.exp(b*t), RGG.mean_degree_list, RGG.mean_nD_list)
    popt = [[-0.5]]
    wx = [s+1 for s in std_degree_list]
    wy = [s+1 for s in std_nD_list] 

    exponential = Model(f)
    data = Data(x, y, wd=wx, we=wy)
    odr = ODR(data, exponential, beta0=popt[0])
    output = odr.run()

    print 'RGG d = ' +str(d) + ' done: gamma = ' + str(output.beta[0]) + ' +/- ' + str(output.sd_beta[0])
    
    ndkhandle, = ndkax.plot(kappa_range_fine_LCC,f(output.beta,kappa_range_fine_LCC), color + '-')
    ndkhandles.append(ndkhandle)
    ndklabels.append('d='+str(d) + ', n=' + str(n) + ', ' + str(boundary))

##################################################
###  LCC scaling ###
##################################################

# ER data and fit
ERE = ex.ERExperiment(kappa_range_LCC, n)
# plt.errorbar(ERE.mean_degree_list,ERE.mean_nD_list,fmt='ro',markersize=3,ecolor='r',xerr=ERE.std_degree_list,yerr=ERE.std_nD_list)
mean_degree_list, std_degree_list, _ = ERE.find_degree_stats()
mean_nD_list, std_nD_list, _ = ERE.find_nD_stats()

x = mean_degree_list
y = mean_nD_list 
wx = [s+1 for s in std_degree_list]
wy = [s+1 for s in std_nD_list] 

# fitting 
# popt = optimize.curve_fit(lambda t,b:np.exp(b*t), RGG.mean_degree_list, RGG.mean_nD_list)
popt = [[-0.5]]
exponential = Model(f)
data = Data(x, y, wd=wx, we=wy)
odr = ODR(data, exponential, beta0=popt[0])
output = odr.run()
ndkhandle, = ndkax.plot(kappa_range_fine_LCC,f(output.beta,kappa_range_fine_LCC),'--')
    
ndkhandles.append(ndkhandle)
ndklabels.append('ER Simulation')

print 'ER sim done', 'gamma = ' + str(output.beta[0]) + ' +/- ' + str(output.sd_beta[0])

## RGG 
for boundary in boundaries:
    if boundary == 's':
        color = 'r'
    elif boundary == 'p':
        color = 'b'
    gammas = []
    gamma_errs = []

    RGG = ex.RGGExperiment(kappa_range_LCC, n, d, shortcut_prob=0, boundary=boundary, num_radii=10)

    mean_degree_list, std_degree_list, _ = RGG.find_degree_stats()
    mean_nD_list, std_nD_list, _ = RGG.find_nD_stats()

    x = mean_degree_list
    y = mean_nD_list

    # plt.errorbar(x, y, fmt='none', yerr=RGG.std_nD_list, xerr=RGG.std_degree_list)
    # fitting 
    # popt = optimize.curve_fit(lambda t,b:np.exp(b*t), RGG.mean_degree_list, RGG.mean_nD_list)
    popt = [[-0.5]]
    wx = [s+1 for s in std_degree_list]
    wy = [s+1 for s in std_nD_list] 

    exponential = Model(f)
    data = Data(x, y, wd=wx, we=wy)
    odr = ODR(data, exponential, beta0=popt[0])
    output = odr.run()

    print 'RGG d = ' +str(d) + ' done: gamma = ' + str(output.beta[0]) + ' +/- ' + str(output.sd_beta[0])
    
    ndkhandle, = ndkax.plot(kappa_range_fine_LCC,f(output.beta,kappa_range_fine_LCC), color + '-.')
    ndkhandles.append(ndkhandle)
    ndklabels.append('d='+str(d) + ', n=' + str(n) + ', ' + str(boundary))


########################

ndkax.set_xlabel('Average Mean Degree $\\langle{k}\\rangle$')
ndkax.set_ylabel('Fraction of Driver Nodes $n_D$')
ndkax.legend(ndkhandles,ndklabels,loc='upper right')
```

```python
# set up plots and axes 
ndkfig = plt.figure(1)
ndkax = ndkfig.add_subplot(1,1,1)
gdfig = plt.figure(2)
gdax = gdfig.add_subplot(1,1,1)

# MODEL
def f(B, x):
    return np.exp(B[0]*x)

# params
kappa_range = [9,10,12,15]
kappa_range_fine = np.arange(8,22,.01)
n = 1000
boundaries = ['s','p']
d_range = [2,3,6,9,15,25]

## ER 
gdhandles = []
ndkhandles = []
gdlegend = []
ndklegend = []

# ER data and fit
ERE = ex.ERExperiment(kappa_range, n)
ERE.to_LCC()
# plt.errorbar(ERE.mean_degree_list,ERE.mean_nD_list,fmt='ro',markersize=3,ecolor='r',xerr=ERE.std_degree_list,yerr=ERE.std_nD_list)

mean_degree_list, std_degree_list, _ = ERE.find_degree_stats()
mean_nD_list, std_nD_list, _ = ERE.find_nD_stats_LCC()

x = mean_degree_list
y = mean_nD_list 
wx = [s+1 for s in std_degree_list]
wy = [s+1 for s in std_nD_list] 

# fitting 
# popt = optimize.curve_fit(lambda t,b:np.exp(b*t), RGG.mean_degree_list, RGG.mean_nD_list)
popt = [[-0.5]]
exponential = Model(f)
data = Data(x, y, wd=wx, we=wy)
odr = ODR(data, exponential, beta0=popt[0])
output = odr.run()
ndkhandle, = ndkax.plot(kappa_range_fine,f(output.beta,kappa_range_fine),'--')
    
ndkhandles.append(ndkhandle)
ndklegend.append('ER Simulation')

ERline, = gdax.plot([min(d_range),max(d_range)],[output.beta[0],output.beta[0]],'k-')
gdax.plot([min(d_range),max(d_range)],[output.beta[0]+2*output.sd_beta[0],output.beta[0]+2*output.sd_beta[0]],'k--')
gdax.plot([min(d_range),max(d_range)],[output.beta[0]-2*output.sd_beta[0],output.beta[0]-2*output.sd_beta[0]],'k--')
gdhandles.append(ERline)
gdlegend.append('Erdos-Renyi')

print 'ER sim done', 'gamma = ' + str(output.beta[0]) + ' +/- ' + str(2*output.sd_beta[0])
    
colors = ['k','c','b','m','g','r']   
markers = ['^','*','s','+','o','v']
    
## RGG 
for boundary in boundaries:
    i = 0 
    if boundary == 's':
        color = 'r'
        BC = 'Solid'
    elif boundary == 'p':
        color = 'b'
        BC = 'Periodic'
    gammas = []
    gamma_errs = []
    for d in d_range:
        RGG = ex.RGGExperiment(kappa_range, n, d, shortcut_prob=0, boundary=boundary, num_radii=10)
        RGG.to_LCC()
        mean_degree_list, std_degree_list, _ = RGG.find_degree_stats()
        mean_nD_list, std_nD_list, _ = RGG.find_nD_stats_LCC()

        x = mean_degree_list
        y = mean_nD_list

        # plt.errorbar(x, y, fmt='none', yerr=RGG.std_nD_list, xerr=RGG.std_degree_list)
        # fitting 
        # popt = optimize.curve_fit(lambda t,b:np.exp(b*t), RGG.mean_degree_list, RGG.mean_nD_list)
        popt = [[-0.5]]
        wx = [s+1 for s in std_degree_list]
        wy = [s+1 for s in std_nD_list] 

        exponential = Model(f)
        data = Data(x, y, wd=wx, we=wy)
        odr = ODR(data, exponential, beta0=popt[0])
        output = odr.run()

        ndkrawhandle = ndkax.errorbar(mean_degree_list,mean_nD_list,xerr=std_degree_list,yerr=std_nD_list,fmt=colors[i]+markers[i],markersize=5)
#         ndkhandles.append(ndkrawhandle)
#         ndklegend.append('Data d='+str(d))

        if boundary == 'p':
            ndkhandle, = ndkax.plot(kappa_range_fine,f(output.beta,kappa_range_fine),colors[i]+'--')
        else:
            ndkhandle, = ndkax.plot(kappa_range_fine,f(output.beta,kappa_range_fine),colors[i]+'-')  
            ndkhandles.append(ndkrawhandle)
            ndklegend.append('d ='+str(d))

        gammas.append(output.beta[0])
        # 95% confidence is 2*standard error for gaussians 
        gamma_errs.append(output.sd_beta[0])
        print str(d) +'th dimension done'
        
        i+=1 
        
    gdhandle = gdax.errorbar(d_range,gammas,yerr=gamma_errs,fmt=color+'o')
    gdhandles.append(gdhandle)
    gdlegend.append(str(BC)+'RGG')

ndkax.set_xlabel('Average Mean Degree $\\langle{k}\\rangle$')
ndkax.set_ylabel('Fraction of Driver Nodes $n_D$')
ndkax.legend(ndkhandles,ndklegend,loc='upper right')

gdax.set_xlabel('Dimension $d$')
gdax.set_ylabel('Scaling Parameter $\\gamma$')
gdax.legend(gdhandles,gdlegend)

# ndkfig.tight_layout()
# gdfig.tight_layout()
ndkfig.savefig('./plots/k_nD_LCC_fits_N_' + str(n) + '_BC_' + boundary + '_one_param.eps',dpi=800)
gdfig.savefig('./plots/gamma_d_LCC_10_BC_' + '_N_' + str(n) + '_one_param.eps',dpi=800)
```
