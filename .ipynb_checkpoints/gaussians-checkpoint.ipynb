# LCC fraction 

```python
n, boundary = 1000, 'g'
d_range = [2,3,6,9,15,25]
kappa_range = [5,10,20,40,60,80]

LCCfig = plt.figure()
LCCax = LCCfig.add_subplot(1,1,1)

### ER ###

frac_in_LCC = []
frac_in_LCC_std = []

ER = ex.ERExperiment(kappa_range, n)

for ensemble in ER.ensembles:
    frac_in_LCC_sample = []
    for sample in ensemble.samples:
        sample.to_LCC()
        # LCC size for each sample 
        frac_in_LCC_sample.append(float(sample.LCC_size)/ER.n)
    # mean over samples of LCC fraction 
    frac_in_LCC.append(np.mean(frac_in_LCC_sample))
    frac_in_LCC_std.append(stats.sem(frac_in_LCC_sample))

mean_degree_list, std_degree_list, _ = ER.find_degree_stats()    

LCCax.errorbar(mean_degree_list, frac_in_LCC, xerr=std_degree_list, yerr=frac_in_LCC_std, label='ER Graph')
print 'ER done'

### RGG ###

for d in d_range:
    frac_in_LCC = []
    frac_in_LCC_std = []

    RGG = ex.RGGExperiment(kappa_range, n, d, shortcut_prob=0, boundary=boundary,num_radii=10)
    
    # find frac stats 
    for ensemble in RGG.ensembles:
        frac_in_LCC_sample = []
        for sample in ensemble.samples:
            sample.to_LCC()
            # LCC size for each sample 
            frac_in_LCC_sample.append(float(sample.LCC_size)/RGG.n)
        # mean over samples of LCC fraction 
        frac_in_LCC.append(np.mean(frac_in_LCC_sample))
        frac_in_LCC_std.append(stats.sem(frac_in_LCC_sample))
    
    # find mean degree stats 
    mean_degree_list, std_degree_list, _ = RGG.find_degree_stats()    
    
    LCCax.errorbar(mean_degree_list, frac_in_LCC, xerr=std_degree_list, yerr=frac_in_LCC_std, label='d = '+str(d))
    print 'd = ' + str(d) + ' done'

legend_handles, legend_strings = LCCax.get_legend_handles_labels()
tenpercent, = LCCax.plot([0,max(kappa_range)],[.9, .9],color='gray',linestyle='-.')
legend_handles.append(tenpercent)
legend_strings.append('90%')

LCCax.set_xlabel('Average Mean Degree $\\langle{k}\\rangle$')
LCCax.set_ylabel('Fraction of Nodes in the LCC')
LCCax.set_ylim([0.45,1.05])
LCCax.legend(legend_handles[0:8], legend_strings[0:8])

# LCCax.set_xticks(range(0, max(kappa_range)+1, 2))
# minor_locator = AutoMinorLocator(2)
# LCCax.xaxis.set_minor_locator(minor_locator)
# LCCax.yaxis.set_minor_locator(minor_locator)
# LCCax.grid(which='both',alpha=0.2)

LCCfig.savefig('./plots/frac_in_LCC_BC_'+str(boundary)+'_N_'+str(n)+'.eps',dpi=800)
```

## stubs 

```python
# find the number of degree-1 nodes in a graph 

def find_stubs(sample):
    
    A = sample.adjacency_dense.copy()
    stub_list = []
    
    for i in range(len(A)):
        if (np.sum(A[:,i]) + np.sum(A[i,:])) == 1:
            stub_list.append(i)
    
    return stub_list
        
def find_isolates(sample):
    
    A = sample.adjacency_dense.copy()
    stub_list = []
    
    for i in range(len(A)):
        if (np.sum(A[:,i]) + np.sum(A[i,:])) == 0:
            stub_list.append(i)
    
    return stub_list   
```

```python
from scipy import stats 

# set up plots and axes 
stubfig = plt.figure(1)
stubax = stubfig.add_subplot(1,1,1)

# params
kappa_range = [20] 
kappa = kappa_range[0]
n = 1000
boundaries = ['g']
d_range = [2,3,6,9,15,25]

## ER 
stubhandles = []
stublabels = []

# ER data
ERE = ex.ERExperiment(kappa_range, n)

i = 0 
color = ['k','y','g','m']

# over kappas 
for ensemble in ERE.ensembles:
    num_stubs_sample = []
    # over samples 
    for sample in ensemble.samples:
        num_stubs_sample.append(len(find_stubs(sample)))
    
    ERmean = np.mean(num_stubs_sample)
    ERerr = stats.sem(num_stubs_sample)

    ERline, = stubax.plot([min(d_range),max(d_range)],[ERmean,ERmean],color[i]+'-')
    stubax.plot([min(d_range),max(d_range)],[ERmean+2*ERerr,ERmean+2*ERerr],color[i]+'--')
    stubax.plot([min(d_range),max(d_range)],[ERmean-2*ERerr,ERmean-2*ERerr],color[i]+'--')
    stubhandles.append(ERline)
    stublabels.append('ER')
    i+=1

    print 'ER sim done', 'mean stubs = ' + str(ERmean) + ' +/- ' + str(ERerr)

## RGG 
for boundary in boundaries:
    color = 'r'
    num_stubs = []
    num_stubs_err = []
    for d in d_range:
        RGG = ex.RGGExperiment(kappa_range, n, d, shortcut_prob=0, boundary=boundary, num_radii=10) 
        for ensemble in RGG.ensembles:
            num_stubs_sample = []
            for sample in ensemble.samples:
                sample.to_LCC()
                num_stubs_sample.append(len(find_stubs(sample)))
        num_stubs.append(np.mean(num_stubs_sample))
        num_stubs_err.append(2*stats.sem(num_stubs_sample))
        print str(d) +'th dimension done'

    stubhandle = stubax.errorbar(d_range,num_stubs,yerr=num_stubs_err,fmt=color+'o')
    stubhandles.append(stubhandle)
    stublabels.append('Gaussian RGG')

stubax.set_xlabel('Dimension $d$')
stubax.set_ylabel('Mean Number of Stubs')
stubax.legend(stubhandles,stublabels)

# ndkfig.tight_layout()
stubfig.tight_layout()
stubfig.savefig('./plots/stubs_Gaussian_N_' + str(n) + '_kappa_' + str(kappa) + '.eps',dpi=800)
```
